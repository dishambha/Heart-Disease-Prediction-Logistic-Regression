# Split the data into feature(x) and target(y)

```python
x = pd.DataFrame(dataset.iloc[:,:-1])
y = pd.DataFrame(dataset.iloc[:,-1])
```

```python
x
```

```python
y
```

# Split the data into training and testing

```python
# Import module to split
from sklearn.model_selection import train_test_split
x_train,x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=23)
```

```python
x_train
```

```python
x_test
```

# Create and train the model

```python
# Import module for fitting
from sklearn.linear_model import LogisticRegression
# create instance (i.e. object) of LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

# Predicting the test set result

```python
y_pred = logmodel.predict(x_test)
```

```python
len(y_pred)
```

```python
y_test.shape
```

# Calculating the accuracy

```python
print('Accuracy : %d', (logmodel.score(x_train,y_train)))
print('Accuracy : %d', (logmodel.score(x_test, y_test)))
```

# Evaluate using confussion Matrix

```python
# confussion matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_train,logmodel.predict(x_train)))
```
